```python
import cv2
import pandas as pd
import os
```

```python
label_path = '../../TrainingDataset/youtube_bounding_box/'
training_label_path = label_path + 'yt_bb_detection_train/youtube_boundingboxes_detection_train.csv'
validation_label_path = label_path + 'yt_bb_detection_validation/youtube_boundingboxes_detection_validation.csv'

save_path = 'D:\\datasets\\TargetBoxDatasets\\YoutubeBoundingBoxes\\videos\\boudingboxes_detection\\'
save_train_path = save_path + 'train\\'
```

```python
from os import listdir
from os.path import isfile, join
onlyfiles = [f[:-4] for f in listdir(save_train_path) if isfile(join(save_train_path, f))]
```

```python
columns = ['youtube_id', 'timestamp_ms', 'class_id', 'class_name', 'object_id', 'object_presence', 
            'xmin', 'xmax', 'ymin', 'ymax']
df_train = pd.read_csv(training_label_path, names=columns) 
# df_val = pd.read_csv(validation_label_path, names=columns) 
```

```python
df_train = df_train[df_train.youtube_id.isin(onlyfiles)]
```

```python
df_train
```

```python
framesize = (255, 255)
fps = 30
fpms = fps/1000

dataset_dic = {}
```

```python
frames_folder_id  = save_path+ '\\frames'
```

```python
framefolder = [f for f in listdir(frames_folder_id)]
```

```python
def store_frames(youtube_id, overwrite=False):
        
    # Create folder to start storing the frames of a youtube video
    video_framespath = save_path+ '\\frames\\'+youtube_id
    
    # If video not already existing in frames folder
    if video_framespath in framefolder or overwrite==True:
    
        video_path = save_train_path+ youtube_id +'.mp4'
        all_timestamp_ms = list(df_train.query('youtube_id == @youtube_id').timestamp_ms.unique())

        cap = cv2.VideoCapture(video_path, ) # Retrieve frames using recorded video

        frame_numbers = [int(timestamp_ms*fpms) for timestamp_ms in all_timestamp_ms]
        
        ret_cap = True
        counter = 0
        frames = {}

        if not os.path.exists(video_framespath):
            os.makedirs(video_framespath)


        while ret_cap: 
            ret_cap, frame = cap.read() # Capture frame-by-frame. If frame is read correctly ret is True
            if ret_cap : 
        #         frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
                frame = cv2.resize(frame, framesize) # resize frame to camera resolution

                if counter in frame_numbers:
                    ts = int(1000*counter/30) # get corresponding timestamps
        #             frames[ts] = frame
                    # Store frame
                    frame_name = video_framespath + f'\\{ts}.jpeg'
                    _ = cv2.imwrite(frame_name, frame)
                counter += 1
        # dataset_dic[youtube_id] = frames
        print(f'Saved frames for {youtube_id} video. Total length : {counter}')
    else : 
        print(f'{video_framespath} already existing')
```

```python
store_frames("ZClABNZVqqw", overwrite=True)
```

```python
for youtube_id in df_train.youtube_id.unique():
    print(f'Storing frames for {youtube_id}')
    store_frames(youtube_id)
```
